##COMPARAÇÃO POR TIPOS DE APARTAMENTOS

```python
apartamentos = df.query('Tipo == "Apartamento"')
apartamentos_por_bairro = apartamentos.groupby('Bairro')[['Valor', 'Condominio', 'IPTU', 'Area', 'Quartos', 'Vagas', 'Suites']].mean().sort_values('Valor', ascending=False)
display(apartamentos_por_bairro)
```

## TOP 20 BAIRROS COM MAIOR E MENOR VALOR MÉDIO DE ALUGUÉL (APARTAMENTOS)

```python
df_apartamento = dados[dados['Tipo'] == 'Apartamento']
df_valor_bairro_ap = (df_apartamento.groupby('Bairro')['Valor'].mean().sort_values(ascending=True).tail(20))

ax = df_valor_bairro_ap.plot(kind='barh', figsize=(14, 8), color='red')
ax.set_title('Top 20 Bairros com Maior Valor Médio de Aluguel (Apartamentos)', fontsize=18)


max_val = df_valor_bairro_ap.max()
ax.set_xlim(0, max_val * 1.1)

ax.set_xticks([])
ax.set_xticklabels([])

ax.xaxis.set_major_locator(plt.MultipleLocator(1000))

for i, v in enumerate(df_valor_bairro_ap):
    ax.text(v + max_val*0.01, i, f'R$ {v:,.0f}'.replace(',', 'X').replace('.', ',').replace('X', '.'),
            va='center')
plt.show()
```

```python
df_apartamento = dados[dados['Tipo'] == 'Apartamento']

df_valor_bairro_ap = (df_apartamento.groupby('Bairro')['Valor'].mean().sort_values(ascending=False).tail(20))

ax = df_valor_bairro_ap.plot(kind='barh', figsize=(14, 8), color='red')
ax.set_title('Top 20 Bairros com Menor Valor Médio de Aluguel (Apartamentos)', fontsize=18)

max_val = df_valor_bairro_ap.max()
ax.set_xlim(0, max_val * 1.1)

ax.set_xticks([])
ax.set_xticklabels([])

ax.xaxis.set_major_locator(plt.MultipleLocator(1000))

for i, v in enumerate(df_valor_bairro_ap):
    ax.text(v + max_val*0.01, i, f'R$ {v:,.0f}'.replace(',', 'X').replace('.', ',').replace('X', '.'),
            va='center')

plt.show()
```

#**CRIANDO COLUNAS CATEGÓRICAS**

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep = ";")
dados.head()
```

```python
dados ['Valor_por_mes'] = dados['Valor'] + dados ['Condominio']
dados.head()
```

```python
dados ['Valor_por_ano'] = dados ['Valor_por_mes'] *12 + dados ['IPTU']
dados.head()
```

#CRIANDO COLUNAS CATEGÓRICAS

```python
dados['Descricao'] = dados ['Tipo'] +  ' em ' + dados  ['Bairro']
dados.head()
```

```python
dados['Descircao'] =  dados ['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                      dados['Quartos'].astype(str)+ ' quartos(s) ' + \
                      ' e ' + dados['Vagas'].astype(str) + '  vaga(s) de garagem  '
dados.head()
```

```python
dados ['Possui_suite']  = dados['Suites'].apply(lambda x:"Sim" if x > 0 else "Não")
dados.head()
```

```python
dados.to_csv('dados_completos_dev.csv', index=False, sep = ';')
```
